## Ridge

```python
from sklearn.linear_model import Ridge

# Random seed
np.random.seed(42)

# Data
X = songs.drop(["track_id", "track_genre", "track_genre_code", "duration_ms", "time_signature"], axis=1)
y = songs["track_genre_code"]
```

```python
songs.shape
```

```python
X.columns
```

```python
y.shape
```

```python
X.head()
```

```python
X.shape
```

```python
# Split data
X_train = X.loc[:70000]
X_val = X.loc[70000:90000]
X_test = X.loc[90000:]

# Y
y_train = y.loc[:70000]
y_val = y.loc[70000:90000]
y_test = y.loc[90000:]

X_train.shape, X_val.shape, X_test.shape
```

```python
for val in y_val[:5]:
    for index, row in songs.iterrows():
        if (row["track_genre_code"] == val):
            genre = row["track_genre"]
            
    print(genre)
```

```python
# Instantiate the model
ridge = Ridge()

# Fit the model
ridge.fit(X_train, y_train)

# Score
ridge.score(X_val, y_val)
```

```python
# Prediction
r_val_pred = ridge.predict(X_val)
```

### Evaluate the Ridge Model

```python
from sklearn.metrics import r2_score, mean_squared_error as mse

ridge_r2 = r2_score(y_val, r_val_pred)
ridge_mse = mse(y_val, r_val_pred)

print("R2 score: ", ridge_r2)
print("MSE score: ", ridge_mse)
```

## Lasso

```python
from sklearn.linear_model import Lasso

# Random seed
np.random.seed(42)

# Instantiate the model
lasso = Lasso()

# Fit the model
lasso.fit(X_train, y_train)

# Score
lasso.score(X_val, y_val)
```

```python
# Predict
l_val_pred = lasso.predict(X_val)
```

```python
# Evaluate
lasso_r2 = r2_score(y_val, l_val_pred)
lasso_mse = mse(y_val, l_val_pred)

print("R2 score: ", lasso_r2)
print("MSE score: ", lasso_mse)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

# Instantiate the model
rfs = RandomForestRegressor()

# Fit the model
rfs.fit(X_train, y_train)

# Score
rfs.score(X_val, y_val)
```

```python
# Predict
rfs_val_pred = rfs.predict(X_val)
```

```python
# Evaluate the model
rfs_r2 = r2_score(y_val, rfs_val_pred)
rfs_mse = mse(y_val, rfs_val_pred)

print("R2 score: ", rfs_r2)
print("MSE score: ", rfs_mse)
```

```python
data = {
    "Ridge": [ridge_r2, ridge_mse],
    "Lasso": [lasso_r2, lasso_mse],
    "Random Forest": [rfs_r2, rfs_mse]
}

scores = pd.DataFrame(data=data, index=["R2", "MSE"])
scores.head()
```

The Random Forest Regressor Model performed better than the other models as it has a lower MSE score and a higher R2 score
