```python
import pandas as pd
import numpy as np
```

```python
songs = pd.read_csv("dataset.csv")
songs.head()
```

```python
# Drop null values
songs = songs.dropna()
```

```python
songs = songs.sample(frac=1).reset_index()

songs.head()
```

```python
songs = songs.drop(["Unnamed: 0", "index", "artists", "album_name", "track_name", "popularity", "explicit"], axis=1)

songs.head()
```

```python
songs.corr()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# Random seed
np.random.seed(42)

# Data
X = songs.drop(["track_id", "track_genre", "duration_ms", "time_signature"], axis=1)
y = songs["track_genre"]

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

# Instantiate the model
knn = KNeighborsClassifier()

# Fit
knn.fit(X_train, y_train)

# Score
knn.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Instantiate the model
rf = RandomForestClassifier()

# Fit
rf.fit(X_train, y_train)

# Score
rf.score(X_test, y_test)
```

```python
y_pred = rf.predict(X_test)
y_pred[:5]
```

```python
# Save the model
import pickle

pickle.dump(rf, open("Music Genre Classifier.pkl", "wb"))
```

```python
# Load model
model = pickle.load(open("Music Genre Classifier.pkl", "rb"))
```

```python
ex = np.array([0.9, 0.5, 1, -8, 0, 0.5, 0.3, 0.1, 0.4, 0.5, 4]).reshape(1,-1)

def predict(ex):
    y_pred = model.predict(ex)
    y_pred = int(y_pred)
    y_pred

    for index, row in songs.iterrows():
        if (row["track_genre_code"] == y_pred):
            print(row["track_genre"])
            break
            
predict(ex)
```
